```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
import datetime 
import calendar 
file = pd.read_csv('./Sales_December_2019.csv')
file.head()
sns.set(rc={'figure.figsize':(15.7,8.27)})
```

```python
file['State'].value_counts()
```

```python
sns.countplot(x='State',data=file)
```

```python
file['City'].value_counts()
```

```python
cities_order=file['City'].value_counts()
cities_order.plot(kind='bar',figsize=(10,5))
```

```python
products_order=file['Product'].value_counts()
products_order.plot(kind='bar',figsize=(15,5))
plt.xlabel('Products')
plt.ylabel('Orders')
plt.title('Mostly purchased products')
```

```python
order=file['Weekday'].value_counts()
order.plot(kind='bar',figsize=(15,10))
plt.xlabel('Weekday')
plt.ylabel('Orders')
plt.title('Day With Mostly purchased products')
```

```python
sns.countplot(x='Weekday',data=file)
```
